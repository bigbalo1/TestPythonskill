The program is inputted with a list of real numbers separated by space.
each element of this list is the radius  of a circle.

write a program that output a list of the areas of these circles, with each area rounded to the 10th decimal place.

if an element in the list is less than zero, it must be skipped when the answer is printed

elements must also be separated by space in the output.
for the constant pi, use  the build-in contant from the math module.
use round() for rounding


```python
from math import pi
def area_of_circle():
  inp = input(print("enter a list containing real numbers separing by commer"))
  lis = inp.split(',')
  result =[]
  for i in lis:
      i = float(i)
      if i >= 0:
         area = pi* i*i
         area = round(area, 10)
         result.append(area)
      else:
         continue


  result = str(result).replace(',',' ')
  result = str(result).replace('[', '')
  result = str(result).replace(']', '')
  print(result)
```
